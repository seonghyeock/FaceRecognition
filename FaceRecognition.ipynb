- 이미지에서 얼굴을 추출하여 저장하는 코드입니다.

```python
import cv2
import dlib
import os
from mtcnn import MTCNN
```

```python
# 이미지 폴더 경로
image_folder_path = r"C:\Users\chosh\Desktop\RecognitionInput"

# 저장할 얼굴 이미지 폴더 경로
output_folder_path = r"C:\Users\chosh\Desktop\RecognitionOutput"
```

```python
detector = MTCNN()

# 이미지 폴더 내 모든 이미지 파일에 대해 처리
for filename in os.listdir(image_folder_path):
    image_path = os.path.join(image_folder_path, filename)
    image = cv2.imread(image_path)
    
    if image is not None:  # Check if the image is not empty
        # BGR을 RGB로 변환
        rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # 얼굴 검출
        faces = detector.detect_faces(rgb_image)
        
        # 각 얼굴에 대해 반복
        for i, face in enumerate(faces):
            # 얼굴 위치 추출
            x, y, w, h = face['box']
            
            # 얼굴 이미지 추출
            face_image = image[y:y+h, x:x+w]
            
            # 추출된 얼굴 이미지 저장
            output_filename = f"{os.path.splitext(filename)[0]}_face_{i}.jpg"
            output_path = os.path.join(output_folder_path, output_filename)
            cv2.imwrite(output_path, face_image)
            
            print(f"얼굴 이미지 저장 완료: {output_path}")
```
